```python
import pandas as pd
import numpy as np


datasetAppA = pd.read_excel("Kuesioner dataset.xlsx",sheet_name='MoodTools')
datasetAppB = pd.read_excel("Kuesioner dataset.xlsx",sheet_name='MyMood')

datasetAppA
```

```python
datasetAppB
```

```python
dataA= datasetAppA.drop(['Timestamp','Usia Anda Sekarang','Pendidikan Terakhir Anda','Pekerjaan Anda Sekarang'], axis=1)
dataB= datasetAppB.drop(['Timestamp','Usia Anda Sekarang','Pendidikan Terakhir Anda','Pekerjaan Anda Sekarang'], axis=1)

dataA['P1'] = dataA['P1'] - 1
dataA['P2'] = dataA['P2'] - 1
dataA['P3'] = dataA['P3'] - 1
dataA['P4'] = dataA['P4'] - 1
dataA['P5'] = dataA['P5'] - 1
dataA['P6'] = dataA['P6'] - 1
dataA['P7'] = dataA['P7'] - 1
dataA['P8'] = dataA['P8'] - 1
dataA['P9'] = dataA['P9'] - 1
dataA['P10'] = dataA['P10'] - 1

dataA
```

```python

dataB['P1'] = dataB['P1'] - 1
dataB['P2'] = dataB['P2'] - 1
dataB['P3'] = dataB['P3'] - 1
dataB['P4'] = dataB['P4'] - 1
dataB['P5'] = dataB['P5'] - 1
dataB['P6'] = dataB['P6'] - 1
dataB['P7'] = dataB['P7'] - 1
dataB['P8'] = dataB['P8'] - 1
dataB['P9'] = dataB['P9'] - 1
dataB['P10'] = dataB['P10'] - 1

dataB
```

```python
dataJenisKelaminA = dataA.groupby('Jenis Kelamin Anda')
dataPerempuanA= dataJenisKelaminA.get_group('Perempuan')
dataLakiA= dataJenisKelaminA.get_group('Laki-Laki')

dataJenisKelaminB = dataB.groupby('Jenis Kelamin Anda')
dataPerempuanB= dataJenisKelaminB.get_group('Perempuan')
dataLakiB= dataJenisKelaminB.get_group('Laki-Laki')

dataPerempuanA = dataPerempuanA.drop(['Peserta','Jenis Kelamin Anda'], axis=1)
dataPerempuanA
```

```python
dataLakiA = dataLakiA.drop(['Peserta','Jenis Kelamin Anda'], axis=1)

dataLakiA
```

```python
dataPerempuanB = dataPerempuanB.drop(['Peserta','Jenis Kelamin Anda'], axis=1)

dataPerempuanB
```

```python
dataLakiB = dataLakiB.drop(['Peserta','Jenis Kelamin Anda'], axis=1)

dataLakiB
```

```python
# Menghitung rata-rata per kolom (column-wise mean)
rataPerempuanA = dataPerempuanA.sum()
rataPerempuanB = dataPerempuanB.sum()

arrPerempuanA = rataPerempuanA.values
arrPerempuanB = rataPerempuanB.values

rataLakiA = dataLakiA.sum()
rataLakiB = dataLakiB.sum()

arrLakiA = rataLakiA.values
arrLakiB = rataLakiB.values
```

```python
import scipy.stats as stats

# Menjalankan uji Wilcoxon Signed-Rank
statistic, p_value = stats.wilcoxon(arrPerempuanA, arrPerempuanB)

# Menampilkan hasil
print("Nilai uji Wilcoxon Signed-Rank:", statistic)
print("Nilai p-value:", p_value)

# Menginterpretasikan hasil
alpha = 0.05  # Tingkat signifikansi
if p_value < alpha:
    print("Terdapat perbedaan yang signifikan antara aplikasi A dan B pada sampel yang sama.")
else:
    print("Tidak cukup bukti untuk menyatakan adanya perbedaan yang signifikan antara aplikasi A dan B pada sampel yang sama.")
```
